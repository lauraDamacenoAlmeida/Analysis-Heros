```python
Q1 = df['APPEARANCES'].quantile(q=0.25)
Q3 = df['APPEARANCES'].quantile(q=0.75)
print(Q1)
print(Q3)
```

```python
amp = Q3-Q1
```

```python
limite_max = Q3+1.5*amp
limite_inf = Q1- 1.5*amp
```

```python

selecao = (df['APPEARANCES']>=limite_inf)&(df['APPEARANCES']<=limite_max)
dados = df[selecao]
dados.boxplot('APPEARANCES')
```
